#Genre 

แสดงผลข้อมูลที่เกี่ยวข้องกับ Genre ของหนังโดยจะใช้จำนวน(No.)เป็นตัวบ่งชี้ว่าประเภทหนังแบบไหนได้รับความนิยมมากที่สุด

```
import matplotlib.pyplot as plt
import numpy as np
fig,axes =  plt.subplots(1,2)
df_mp = pd.read_csv('Rotten_tomatoes.csv')
df_IMDb = pd.read_csv('IMDb.csv')

sumInGenre = df_IMDb['Genre'].value_counts().sum()
df_IMDb['Genre'].value_counts().head(10).plot(ax=axes[0],kind='bar' , title = ("Top 10 Of IMDb Genre"), figsize=(15,10), color='#f3ce13')
axes[0].set_xticklabels(df_IMDb['Genre'].value_counts().head(10).index.tolist(),rotation = 30,ha='right')
axes[0].set_ylabel('No.', fontsize=16)
axes[0].set_xlabel('Genre', fontsize=16)

sumInGenre = df_mp['Genre'].value_counts().sum()
df_mp['Genre'].value_counts().head(10).plot(ax=axes[1], kind='bar', title = ("Top 10 Of Rotten Tomatoes Genre") , figsize=(15,10), color='#7a2f2f')
axes[1].set_xticklabels(df_mp['Genre'].value_counts().head(10).index.tolist(),rotation = 30,ha='right')
axes[1].set_ylabel('No.', fontsize=16)
axes[1].set_xlabel('Genre', fontsize=16)
```

#Rating

แสดงภาพข้อมูลเกี่ยวกับคะแนนของภาพยนต์โดยจะมีจำนวน(No.)เป็นตัววัดว่าภาพยนต์คะแนนที่เท่าไหร่มีกี่จำนวน

```
import matplotlib.pyplot as plt
import numpy as np
fig,axes =  plt.subplots(1,2)
df_mp = pd.read_csv('IMDb.csv')


sumInGenre = df_mp['Rating'].value_counts().sum()
df_mp['Rating'].value_counts().plot(ax=axes[0],kind='bar', title = ("Top 10 Of IMDb Rating") , figsize=(15,10), color='#f3ce13')
axes[0].set_xticklabels(df_mp['Rating'].value_counts().head(10).index.tolist(),rotation = 30,ha='right')
axes[0].set_ylabel('No.', fontsize=16)
axes[0].set_xlabel('Rating', fontsize=16)

df_mp = pd.read_csv('Rotten_tomatoes.csv')
sumInGenre = df_mp['Rating'].value_counts().sum()
df_mp['Rating'].value_counts().plot(ax=axes[1],kind='bar', title = ("Top 10 Of Rotten Tomatoes Rating") , figsize=(15,10), color='#7a2f2f')
plt.xticks(rotation = 30,ha='right')
plt.ylabel('No.', fontsize=16)
plt.xlabel('Rating', fontsize=16)
```

#Rate

แสดงภาพข้อมูลเกี่ยวกับ Rate ของหนังที่ได้รับความนิยมมากที่สุด

```
import matplotlib.pyplot as plt
import numpy as np
fig,axes =  plt.subplots(1,2)
df_mp = pd.read_csv('IMDb.csv')
sumInGenre = df_mp['R'].value_counts().sum()

df_mp['R'].value_counts().plot(ax=axes[0],kind='bar', title = ("IMDb Rate") , figsize=(15,10), color='#f3ce13')
axes[0].set_xticklabels(df_mp['R'].value_counts().index.tolist(),rotation = 30,ha='right')
axes[0].set_ylabel('No.', fontsize=16)
axes[0].set_xlabel('R', fontsize=16)

df_mp = pd.read_csv('Rotten_tomatoes.csv')
sumInGenre = df_mp['R'].value_counts().sum()
df_mp['R'].value_counts().plot(ax=axes[1],kind='bar', title = ("Rotten Tomatoes Rate") , figsize=(15,10), color='#7a2f2f')
plt.xticks(rotation = 30,ha='right')
plt.ylabel('No.', fontsize=16)
plt.xlabel('R', fontsize=16)
```
