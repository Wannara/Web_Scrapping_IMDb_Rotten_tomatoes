# Rotten Tomatoes การทำ Web scrapping และการจัดรูปแบบให้อยู่ในรูป CSV File เพื่อนำข้อมูลไปใช้งาน

```
from bs4 import BeautifulSoup
import requests
import urllib.request
import re
```

# Titles, Years, Ratings




```
url = "https://editorial.rottentomatoes.com/guide/best-netflix-movies-to-watch-right-now/"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')

links = [a.attrs.get('href') for a in soup.select('div.article_movie_title a')]

title = [a.get_text() for a in soup.select('div.article_movie_title a')]

years = list(map(lambda x: int(x[1:-1]), [a.get_text() for a in soup.select('span.start-year')]))

ratings = list(map(lambda y:float(y.strip("%"))/10, [a.get_text() for a in soup.select('span.tMeterScore')]))
```

# Grene

```
genre = []

for url in links:

    r = requests.get(str(url))
    soup = BeautifulSoup(r.content, "html.parser")
    data = soup.find("div", {"class":"genre"})

    try:
        g = data.text.split()
        ge = [x.strip(",") for x in g]

        if 'and' in ge:
            ge.remove("and")
        if 'sci' in ge and 'fi' in ge:
            ge.remove("sci")
            ge.remove("fi")
            ge.append("sci-fi")

        for j in range(len(ge)):
            
            ge[j] = ge[j].title()
    
    except:
        ge = []

    genre.append(ge)
```

# Rate

```
rate = []
for url in links:
    response = requests.get(str(url))
    soup = BeautifulSoup(response.text, 'lxml')
    data = [a.get_text().strip().split(",")[0] for a in soup.select("div.meta-value")]
    if "R" in data[0] or "PG" in data[0]:
        rate.append(data[0])
    else:
        rate.append("Unknow")
print(rate)
```

```
title.reverse()
years.reverse()
ratings.reverse()
genre.reverse()
rate.reverse()

print(title)
print(years)
print(ratings)
print(genre)
print(rate)
```
