# Predicton ค่า Rating ของ IMDb และ Rotten tomatoes

```
#Prediction Rating of IMDb
def prediction_IMDB(genre,r):
  genre = genre.lower()
  gen_coef = ['music','fantasy','history','biography','drama','action','western','crime']
  r_coef = ['g','pg','approved','r','not rated','pg-13']
  a = 0
  b = 0
  c = 0
  d = 0
  if(genre in gen_coef):
    a = 1
  if(genre == 'crime'):
    b = 1
  if(r in r_coef):
    c = 1
  if(r == 'pg-13'):
    d = 1
  predict = 0.0833 * a + 0.0915 * b + 0.1215 * c +0.1601 * d + 8.3069
  return predict

#Prediction Rating of Rotten tomatoes
def prediction_Rotten_Tomato(genre,r):
  genre = genre.title()
  r = r.upper()
  gen_coef1 = ['War','Romance','Mystery','Thriller','Horror','Sports','Fitness','Sci-Fi,Action','Drama','Comedy','Crime','History','Other','Family','Kids','Adventure','Fantasy','Animation','Documentary','Biography','Western']
  gen_coef2 = ['Drama','Comedy','Crime','History','Other','Family','Kids','Adventure','Fantasy','Animation','Documentary','Biography','Western']
  gen_coef3 = ['Documentary','Biography','Western']
  r_coef = ['PG-13','PG']
  a = 0
  b = 0
  c = 0
  d = 0
  if(genre in gen_coef1):
    a = 1
  if(genre in gen_coef2):
    b = 1
  if(genre in gen_coef3):
    c = 1
  if(r in r_coef):
    d = 1

  predict = 0.2185 * a + 0.0561 * b + 0.2125 * c + 0.1023 * c + 9.1345
  return predict


# Main for prediction
def main():
  select = int(input('Selected IMDB press 1 or Rotten Tomatoes press 2 : '))
  
  if(select == 1):
    print("IMDB")
    print('Example Genre: music,fantasy,history,biography,drama,action,western,crime')
    genre = input('Input Genre : ')
    print('Example Rate:g,pg,approved,r,not rated,pg-13')
    r = input('Input Rate: ')
    print("%.2f"%prediction_IMDB(genre,r))
  else:
    print("Rotten Tomatoes")
    print('Example Genre: War,Romance,Mystery,Thriller,Horror,Sports,Fitness,Sci-Fi,Action,Drama,Comedy,Crime,History,Other,Family,Kids,Adventure,Fantasy,Animation,Documentary,Biography,Western')
    genre = input('Input Genre : ')
    print('Example Rate: g,pg,approved,r,not rated,pg-13')
    r = input('Input Rate: ')
    result = prediction_Rotten_Tomato(genre,r)
    result *= 10
    print("%.2f"%result)

main()
  
```
